# **Imports**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.datasets import make_classification

import warnings
warnings.filterwarnings("ignore")

enable_iterative_imputer
```

# **Creating Dataset**

```python
num_samples = 1000
X_numeric, y = make_classification(n_samples=num_samples, n_features=3, 
                                    n_informative=3, n_redundant=0, 
                                    n_clusters_per_class=1, random_state=17)

X_integer = np.random.randint(1, 100, size=(num_samples, 2))

categories = ['A', 'B', 'C', 'D']
X_categorical = np.random.choice(categories, size=(num_samples, 2))

df = pd.DataFrame(X_numeric, columns=['cont_1', 'cont_2', 'cont_3'])
df['int_1'] = X_integer[:, 0]
df['int_2'] = X_integer[:, 1]
df['cat_1'] = X_categorical[:, 0]
df['cat_2'] = X_categorical[:, 1]
df['target'] = y  
```

# **Function for finding optimal k for KNN**

```python
def find_optimal_k(X_train, y_train, cv_folds=5):
    k_values = list(range(1, 30))  
    mean_scores = []

    for k in k_values:
        imputer = KNNImputer(n_neighbors=k)
        X_imputed = imputer.fit_transform(X_train) 
        model = KNeighborsClassifier(n_neighbors=5)
        scores = cross_val_score(model, X_imputed, y_train, cv=cv_folds, scoring='accuracy')
        mean_scores.append(np.mean(scores))

    optimal_k = k_values[np.argmax(mean_scores)]
    print(f"Optimal k found: {optimal_k}")
    return optimal_k
```

## **Dividing data for 2 types**

```python
numeric_features = ['cont_1', 'cont_2', 'cont_3', 'int_1', 'int_2']
categorical_features = ['cat_1', 'cat_2']
```

## **Creating a Pipeline**

```python
def create_pipeline(imputer, cat_imputer):
    numeric_transf = Pipeline(steps=[
        ('imputer', imputer),
        ('scaler', StandardScaler())
    ])

    categorical_transf = Pipeline(steps=[
        ('imputer', cat_imputer),
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ])

    return ColumnTransformer(transformers=[
        ('numeric', numeric_transf, numeric_features),
        ('categorical', categorical_transf, categorical_features)
    ])
```

## **Training different models**

```python
def train_models(X_train, y_train, X_test, y_test, strategy_name, col_prep):
    results = {}

    common_params = {
        "n_estimators": 100,
        "learning_rate": 0.1,
        "max_depth": 6
    }

    models = {
        "XGBoost": XGBClassifier(**common_params, use_label_encoder=False, eval_metric='mlogloss'),
        "LightGBM": LGBMClassifier(**common_params),
        "CatBoost": CatBoostClassifier(**common_params, verbose=0)
    }

    for name, model in models.items():
        pipeline = Pipeline(steps=[
            ('col_prep', col_prep),
            ('model', model)
        ])

        pipeline.fit(X_train, y_train)
        y_pred = pipeline.predict(X_test)

        results[name] = {
            "Accuracy": accuracy_score(y_test, y_pred),
            "Precision": precision_score(y_test, y_pred, average='macro', zero_division=0),
            "Recall": recall_score(y_test, y_pred, average='macro', zero_division=0),
            "F1 Score": f1_score(y_test, y_pred, average='macro', zero_division=0)
        }
        print(f"{strategy_name} - {name}: {results[name]}")

    return results
```

## **Loading and training a Baseline**

```python
X = df[numeric_features + categorical_features]
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)

# Базова модель без пропущених значень
col_prep_baseline = create_pipeline(SimpleImputer(strategy='median'), SimpleImputer(strategy='most_frequent'))
baseline_results = train_models(X_train, y_train, X_test, y_test, "Baseline", col_prep_baseline)
```

## **Creating a dataset with missing values**

```python
df_missing = df.copy()
missing_mask = np.random.rand(*df_missing.shape) < 0.5
df_missing.where(~missing_mask, np.nan, inplace=True)
X_missing = df_missing[numeric_features + categorical_features]
X_train_miss, X_test_miss, y_train_miss, y_test_miss = train_test_split(X_missing, y, test_size=0.2, random_state=17)
```

## **Strategies of Imputation**|

```python
optimal = find_optimal_k(X_train_miss[numeric_features], y_train_miss, cv_folds=5)

strategies = {
    "Mean": SimpleImputer(strategy='mean'),
    "Median": SimpleImputer(strategy='median'),
    "Most Frequent": SimpleImputer(strategy='most_frequent'),
    "KNN Imputer": KNNImputer(n_neighbors=optimal),
    "Iterative Imputer": IterativeImputer(max_iter=15, random_state=17)
}

strategies_cat = {
    "Constant": SimpleImputer(strategy='constant', fill_value="missing"),
    "Most Frequent": SimpleImputer(strategy='most_frequent')
}
```

# **Train models**

```python
final_results = {"Baseline": baseline_results}

for strategy_name, imputer in strategies.items():
    for cat_strategy_name, cat_imputer in strategies_cat.items():
        col_prep = create_pipeline(imputer, cat_imputer)
        results = train_models(X_train_miss, y_train_miss, X_test_miss, y_test_miss,
                               f"{strategy_name} + {cat_strategy_name}", col_prep)
        final_results[f"{strategy_name} + {cat_strategy_name}"] = results
```

# **Visualization**

```python
fig, ax = plt.subplots(figsize=(12, 6))
metrics = ["Accuracy", "Precision", "Recall", "F1 Score"]
strategies_list = list(final_results.keys())

for metric in metrics:
    metric_values = [np.mean([model[metric] for model in final_results[strategy].values()]) for strategy in
                     strategies_list]
    ax.plot(strategies_list, metric_values, marker='o', label=metric)

ax.set_title("Comparison of Imputation Strategies")
ax.set_xlabel("Imputation Strategy")
ax.set_ylabel("Score")
ax.legend()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
metrics = ["Accuracy", "Precision", "Recall", "F1 Score"]
models = ["XGBoost", "LightGBM", "CatBoost"]
strategies_list = list(final_results.keys())

fig, axes = plt.subplots(1, len(metrics), figsize=(15, 5), sharey=True)

for i, metric in enumerate(metrics):
    ax = axes[i]
    
    for strategy in strategies_list:
        metric_values = [final_results[strategy][model][metric] for model in models]
        ax.plot(models, metric_values, marker='o', label=strategy)

    ax.set_title(metric)
    ax.set_xlabel("Model")
    ax.set_ylabel("Score")
    ax.legend(fontsize='small')

plt.suptitle("Comparison of Model Performance Across Imputation Strategies", fontsize=14)
plt.show()
```
