# **Imports**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from xgboost import XGBClassifier

import warnings
warnings.filterwarnings("ignore")

enable_iterative_imputer
```

# **Generation of dataset**

```python
np.random.seed(17)
n_samples = 1000
n_features = 10
n_classes = 4

X = np.random.randn(n_samples, n_features) * 10
y = np.random.randint(0, n_classes, size=n_samples)
```

## **Adding missing values and converting into DataFrame**

```python
missing_rate = 0.9
mask = np.random.rand(*X.shape) < missing_rate
X_missing = X.copy()
X_missing[mask] = np.nan


columns = [f'feature_{i}' for i in range(n_features)]
X_missing_df = pd.DataFrame(X_missing, columns=columns)
y_df = pd.Series(y, name='target')
```

### **Train-test split**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_missing_df, y_df, test_size=0.2, random_state=17, stratify=y_df
)
```

## **Imputation Strategies**

```python
strategies = {}

# Drop Missing
X_train_dropped = X_train.dropna()
y_train_dropped = y_train.loc[X_train_dropped.index]
strategies["Drop Missing"] = (X_train_dropped, y_train_dropped)
```

```python
for strat_name, imputer in [
    ("Mean Imputation", SimpleImputer(strategy='mean')),
    ("Median Imputation", SimpleImputer(strategy='median')),
    ("Most Frequent Imputation", SimpleImputer(strategy='most_frequent')),
    ("KNN Imputation", KNNImputer(n_neighbors=5)),
    ("Iterative Imputation", IterativeImputer(max_iter=10, random_state=17)),
    ("Constant Imputation (Zero)", SimpleImputer(strategy='constant', fill_value=0)),
    ("Constant Imputation (Mean - 1 Std)", SimpleImputer(strategy='constant', fill_value=(np.nanmean(X_train) - np.nanstd(X_train)))),
    ("Constant Imputation (Mean + 1 Std)", SimpleImputer(strategy='constant', fill_value=(np.nanmean(X_train) + np.nanstd(X_train))))
]:
    X_train_filled = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
    strategies[strat_name] = (X_train_filled, y_train)

strategies["Forward Fill"] = (X_train.fillna(method='ffill'), y_train)
strategies["Backward Fill"] = (X_train.fillna(method='bfill'), y_train)
strategies["Mode Imputation"] = (X_train.apply(lambda col: col.fillna(col.mode()[0]) if col.dtype != 'O' else col, axis=0), y_train)
strategies["Interpolation"] = (X_train.interpolate(method='linear'), y_train)
strategies["Polynomial Interpolation"] = (X_train.interpolate(method='polynomial', order=2), y_train)
```

## **Creating a Baseline**

```python
X_baseline_train, X_baseline_test, y_baseline_train, y_baseline_test = train_test_split(
    X, y, test_size=0.2, random_state=17, stratify=y
)
```

## **Hyperparameters**

```python
xgb_params = {
    "n_estimators": 100,
    "learning_rate": 0.001,
    "max_depth": 3,
    "random_state": 17,
    "objective": "multi:softmax",
    "num_class": n_classes
}
```

## **Baseline Model Training**

```python
baseline_model = XGBClassifier(**xgb_params)
baseline_model.fit(X_baseline_train, y_baseline_train)
y_baseline_pred = baseline_model.predict(X_baseline_test)
baseline_accuracy = accuracy_score(y_baseline_test, y_baseline_pred)

print(f"Baseline Accuracy (No Missing Values): {baseline_accuracy:.4f}")
```

```python
baseline_metrics = {
    "Accuracy": accuracy_score(y_baseline_test, y_baseline_pred),
    "Precision": precision_score(y_baseline_test, y_baseline_pred, average='macro', zero_division=0),
    "Recall": recall_score(y_baseline_test, y_baseline_pred, average='macro', zero_division=0),
    "F1 Score": f1_score(y_baseline_test, y_baseline_pred, average='macro', zero_division=0)
}

print("\nBaseline (No Missing Values) Metrics:")
for metric, value in baseline_metrics.items():
    print(f"{metric}: {value:.4f}")
    
    
results = {"Baseline (No Missing Values)": baseline_metrics}
```

## **Different imputation strategies training**

```python
X_test_filled = pd.DataFrame(SimpleImputer(strategy='mean').fit_transform(X_test), 
                             columns=X_test.columns, index=X_test.index)

for strategy, (X_train_filled, y_train_filled) in strategies.items():
    if X_train_filled.shape[0] != y_train_filled.shape[0]:
        print(f"Skipping {strategy} due to row mismatch")
        continue

    model = XGBClassifier(**xgb_params)
    model.fit(X_train_filled, y_train_filled)

    y_pred = model.predict(X_test_filled)
    
    metrics = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average='macro', zero_division=0),
        "Recall": recall_score(y_test, y_pred, average='macro', zero_division=0),
        "F1 Score": f1_score(y_test, y_pred, average='macro', zero_division=0)
    }

    results[strategy] = metrics
```

## **Visualization**

```python
metrics_names = ["Accuracy", "Precision", "Recall", "F1 Score"]

for metric in metrics_names:
    plt.figure(figsize=(12, 6))
    values = [results[strategy][metric] for strategy in results.keys()]
    sns.barplot(x=list(results.keys()), y=values, palette="viridis")

    plt.xticks(rotation=90)
    plt.ylabel(metric)
    plt.xlabel("Imputation Strategy")
    plt.axhline(y=baseline_metrics[metric], color='r', linestyle='--', label="Baseline")
    plt.legend()

    plt.title(f"XGBoost Performance: {metric}")
    plt.show()
```

```python
print("\nFinal Results:")
for strat, metrics in results.items():
    print(f"\n{strat}:")
    for metric, value in metrics.items():
        print(f"  {metric}: {value:.4f}")
```
